## Feature Engineering

The target variable was encoded numerically to facilitate correlation analysis and machine learning modeling.


```python
df["Heart Disease Numeric"]=df["Heart Disease"].map({"Absence": 0, "Presence": 1})
numeric_columns=df.select_dtypes(include=["int64","float64"])
corr_matrix=numeric_columns.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

```

```python
corr_with_heartdisease = corr_matrix['Heart Disease Numeric'].sort_values(ascending=False)
print(corr_with_heartdisease)
```

```python
corr_with_heartdisease = corr_matrix["Heart Disease Numeric"].sort_values()
corr_with_heartdisease.plot(kind="barh", figsize=(8,6))
plt.title("Correlation with Heart Disease")
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score,classification_report,accuracy_score , confusion_matrix
from sklearn.model_selection import train_test_split


```

## Machine Learning Models

Multiple classification models were trained and evaluated to predict heart disease presence.


```python
x=df.drop(["Heart Disease" ,"Heart Disease Numeric"] ,axis=1)
y=df["Heart Disease Numeric"]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train ,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)

```

### Logistic Regression

Logistic Regression was selected as a baseline model due to its interpretability and effectiveness in medical datasets. Features were standardized prior to training.


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score,classification_report,accuracy_score , confusion_matrix
logreg_model=LogisticRegression(max_iter=1000)
logreg_model.fit(x_train_scaled, y_train)
```

```python
y_predicted=logreg_model.predict(x_test_scaled)
```

```python
print("Accuracy:",accuracy_score(y_test,y_predicted))
print("Classification:",classification_report(y_test,y_predicted))
print("Confusion_matrix:",confusion_matrix(y_test,y_predicted))
```

```python
roc_lr= roc_auc_score(y_test, logreg_model.predict_proba(x_test_scaled)[:,1])
```

```python
print(roc_lr)
```

```python
cm = confusion_matrix(y_test, y_predicted)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

### Decision Tree Classifier

A Decision Tree model was trained to capture non-linear relationships. However, single trees are prone to overfitting, particularly on small datasets.


```python
from sklearn.tree  import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score,classification_report ,accuracy_score ,confusion_matrix
decisiontree_model= DecisionTreeClassifier(max_depth=4 ,min_samples_leaf=200,random_state=42)
decisiontree_model.fit(x_train,y_train)

```

```python
y_predicted=decisiontree_model.predict(x_test)
```

```python
print("Accuracy:",accuracy_score(y_test,y_predicted))
print("Classification:",classification_report(y_test,y_predicted))
print("Confusion_matrix:",confusion_matrix(y_test,y_predicted))
```

```python
roc_dt= roc_auc_score(y_test, decisiontree_model.predict_proba(x_test)[:,1])
```

```python
print(roc_dt)
```

### Random Forest Classifier

Random Forest was used as an ensemble approach to improve stability and performance by averaging multiple decision trees.


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score,classification_report,accuracy_score,confusion_matrix
randomforest_model=RandomForestClassifier(max_depth=6 ,min_samples_leaf=5,random_state=42 ,n_estimators=200)
randomforest_model.fit(x_train,y_train)
```

```python
y_predicted=randomforest_model.predict(x_test)
```

```python
print("Accuracy:",accuracy_score(y_test,y_predicted))
print("Classification:",classification_report(y_test,y_predicted))
print("Confusion_matrix:",confusion_matrix(y_test,y_predicted))
```

```python
roc_rf= roc_auc_score(y_test, randomforest_model.predict_proba(x_test)[:,1])
```

```python
print(roc_rf)
```

## Model Evaluation and Comparison

Models were evaluated using ROC-AUC, which measures the ability to discriminate between patients with and without heart disease.


```python
results = pd.DataFrame({
    "Model": ["Logistic Regression", "Decision Tree", "Random Forest"],
    "ROC-AUC": [roc_lr, roc_dt, roc_rf]
})

results
```

Logistic Regression achieved the highest ROC-AUC, followed closely by Random Forest. The Decision Tree performed no better highlighting its instability on this dataset.


## Conclusion

This project demonstrated that heart disease can be effectively predicted using clinical and diagnostic features. Logistic Regression provided the best balance between performance and interpretability, achieving a ROC-AUC of approximately 0.95.

Key predictors included Thallium test results, number of affected vessels, exercise-induced angina, and maximum heart rate.

## Limitations and Future Work

- Additional validation using cross-validation is recommended
